```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
from pylab import rcParams
%matplotlib inline
%config InlineBackend.figure_format='svg'
plt.style.use('fivethirtyeight')
```

```python
data=pd.read_csv('data/creditcard.csv')
data.head(10)
```

```python
x=data['Class'].value_counts()
x
```

```python
x.plot(kind="bar")
plt.show()
```

```python
x.plot(kind="bar",logy=True)
plt.show()
```

```python
a=data.loc[data['Class']==0,'V1'].values
```

```python
b=data.loc[data['Class']==1,'V1'].values
```

```python
fig,ax=plt.subplots(nrows=1,ncols=2)
fig.set_size_inches(10, 6)
fig.subplots_adjust(wspace=0.3)
ax1,ax2=ax.flatten()
ax1.hist(a,bins=20,density=True,alpha=0.5,label='Class 0',color='gray')
ax2.hist(b,bins=20,density=True,color='red',alpha=0.5,label='Class 1')
ax1.legend(fontsize = 8)
ax2.legend(fontsize = 8)
ax1.set_xlabel('Class')
ax2.set_xlabel('Class')
plt.show()
```
